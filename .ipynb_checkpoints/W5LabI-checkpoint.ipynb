```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\fabia\Documents\Ironhack ALL\Coursework\W5\Dataset\amz_uk_price_prediction_dataset.csv")
df.columns
```

```python
# Frequency Table for Product Categories
category_counts = df['category'].value_counts()
display(category_counts)
```

```python
top_5_categories = category_counts.head(5)
display(top_5_categories)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_5_categories.index, top_5_categories.values)
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for Proportions of Top 5 Categories with Matplotlib
# Pie Chart for Proportions of Top 5 Categories with Custom Colors
plt.figure(figsize=(8, 8))
custom_colors_pie = ['#00BFBF', '#7FFFD4', '#F0FFFF', '#0000FF', '#6495ED']
plt.pie(top_5_categories, labels=top_5_categories.index, autopct='%1.1f%%', startangle=90, colors=custom_colors_pie)
plt.title('Proportions of Top 5 Categories')
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for Proportions of Top 5 Categories with Adjusted Labels
plt.figure(figsize=(10, 10))
explode = [0.1] + [0] * (len(top_5_categories) - 1)  # Explode the first slice for emphasis
custom_colors_pie = ['#00BFBF', '#7FFFD4', '#F0FFFF', '#0000FF', '#6495ED']

plt.pie(
    top_5_categories,
    labels=top_5_categories.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=custom_colors_pie,
    explode=explode,
    labeldistance=1.2  # Adjust label distance to prevent overlap
)
plt.title('Proportions of Top 5 Categories', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]  # [0] selects the first mode in case there are multiple

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

```python
# Measures of Dispersion
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()

q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr_price = q3 - q1

print(f"Variance: {variance_price}")
print(f"Standard Deviation: {std_dev_price}")
print(f"Range: {range_price}")
print(f"Interquartile Range: {iqr_price}")
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['price'], bins=30, edgecolor='k', alpha=0.7)
plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'], vert=False, patch_artist=True)
plt.title("Box Plot of Product Prices")
plt.xlabel("Price")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import skew, kurtosis

# Measures of Centrality
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")

# Measures of Dispersion
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()

q1_rating = df['stars'].quantile(0.25)
q3_rating = df['stars'].quantile(0.75)
iqr_rating = q3_rating - q1_rating

print(f"Variance of Rating: {variance_rating}")
print(f"Standard Deviation of Rating: {std_dev_rating}")
print(f"Interquartile Range (IQR): {iqr_rating}")

# Shape of the Distribution
skewness_rating = skew(df['stars'])
kurtosis_rating = kurtosis(df['stars'])

print(f"Skewness: {skewness_rating}")
print(f"Kurtosis: {kurtosis_rating}")

# Visualization
plt.figure(figsize=(8, 6))
plt.hist(df['stars'], bins=10, edgecolor='k', alpha=0.7)
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
